## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

### Ans 5

# first we define a function in order to calculate a revenue of each item as per the company charges

```python
def rev(x):
    if (x>600):
        rev=x*.25
        return rev
    elif (x<=600 and x>350):
        rev=x*.15
        return rev
    elif (350>=x>100):
        rev=x*.10
        return rev
    else:
        rev=x*.05
        return rev
```

# now we create a list variable 'revenue' in order to calulate the revenue of each item and store it . Also add a new column "Revenue" in a dataset "ECom_Data" having brand revenue data  in it

```python
revenue=[rev(i) for i in ECom_Data['discounted_price']]
ECom_Data['Revenue']=revenue
```

#  since we have to calculate the Total Revenue generated by the E-Commerce company over all orders placed, we do the sum  of column 'Revenue'

```python
total=ECom_Data['Revenue'].sum()
print('Total Revenue generated is \u20B9',round(total),sep='')
```

# Answer 05:- Total Revenue generated is ₹2217487

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately.  (4 marks) 

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days. Also, show Month and Year on x-axis.

### Ans 9 a)

# first we need to change the Order_Date column data type from object datatype to datetime datatype

```python
ECom_Data['Order_Date']=pd.to_datetime(ECom_Data['Order_Date'],dayfirst=True)
ECom_Data.info()
```

# datetime() is used in order to convert object datatype to datetime datatype.
# In above info we can see that the Order_Date column datatype has been changed in datetime datatype

# As Month and Year are required on X-axis so we need to extract month & year from Order_Date and create a new column name month_year in the dataframe ECom_Data and store month and year in that column

```python
ECom_Data['month_year']=ECom_Data['Order_Date'].dt.strftime('%b-%Y')
ECom_Data.sort_values('Order_Date',inplace=True)
```

# In dt.strftime() is used in order to extract day,month and year. Also '%b' would give us a month and '%Y' would give us a year

# now we have all the dataset in order to create a lineplot i.e month_year(x-axis),revenue(brand_revenue) on y-axis and region will be taken as hue paramter

```python
plt.figure(figsize=(15,7))
sns.lineplot(data=ECom_Data,x='month_year',y='Revenue',hue='Region',errorbar=('ci', False),style='Region');
```
